```python
import pandas as pd
import glob

# Path to your folder where CSVs are stored
path = r"C:\Users\kisho\Downloads\Big projects\cool\Federal Spending data\Files\*.csv"                                                                 

# Get all CSV file paths
files = glob.glob(path)

print("Files detected:")
for f in files:
    print(f)

# Read and combine everything
df_list = []
for f in files:
    print("Loading:", f)
    df = pd.read_csv(f, low_memory=False)
    df_list.append(df)

# Concatenate all into one big DataFrame
big_df = pd.concat(df_list, ignore_index=True)

print("\nCombined shape (rows, columns):", big_df.shape)

# Save as a single big CSV
big_df.to_csv(r"C:\Users\kisho\Downloads\Big projects\cool\Federal Spending data\all_federal_contracts_2020_2025.csv", index=False)

print("\nDONE! Saved as all_federal_contracts_2020_2025.csv")
```
